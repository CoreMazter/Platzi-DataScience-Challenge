# First we import the needed resources

````python
import pandas as pd #for reading the csv and working with it
````

# Then, read the .csv
### In this case, we have some lines with errors, so i decided to just skip them using the ```error_bad_lines``` option

````python
books = pd.read_csv('books.csv',error_bad_lines=False)
books
````

# Which book has the most text_reviews_count?

### To get this, we have to order the dataframe and consult the first value

````python
books.sort_values('text_reviews_count',ascending=False).iloc[0]
````

# Compare top 10 in a graph
### Now we can see, the "Twiligth" book followed by the "The Book Thief" book are considerably above all the others

````python
top_10_text_reviews_count = books.sort_values('text_reviews_count',ascending=False).iloc[0:10].plot.bar(x="title",y="text_reviews_count")
top_10_text_reviews_count
````

# Get the best rated book

````python
max_rating_mask = books['average_rating']==5.0
best_rateds = books[max_rating_mask]
best_rateds
````

# Get the best rated book with most text reviews

````python
best_rated_and_reviewed = best_rateds.sort_values('text_reviews_count',ascending = False)
best_rated_and_reviewed.iloc[0:7].plot.bar(x="title",y="text_reviews_count")
````
